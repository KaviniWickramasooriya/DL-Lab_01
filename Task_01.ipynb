```python
import numpy as np

#Create a random 4x4 array from an exponential distribution
arr = np.random.exponential(size=(4,4))
print("4*4 Exponential Array:\n", arr)
```

```python
import matplotlib.pyplot as plt

#Generate arrays
exponential_arr = np.random.exponential(size=(100000,1))
uniform_arr = np.random.uniform(size=(100000,1))
normal_arr = np.random.normal(size=(100000,1))

#Plot histograms
plt.hist(exponential_arr, density=True, bins=100, histtype="step", color="red", label="Exponential")
plt.hist(uniform_arr, density=True, bins=100, histtype="step", color="blue", label="Uniform")
plt.hist(normal_arr, density=True, bins=100, histtype="step", color="green", label="Normal")
plt.axis([-2.5, 2.5, 0, 1.1])
plt.legend(loc="upper left")
plt.title("Random Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

#Meshgrid
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x,y)
Z = X**2+Y**2

#Plot
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection="3d")
ax.plot_surface(X, Y, Z, cmap="viridis")

ax.set_title("3D Plot: Z = X^2 + Y^2")
ax.set_xlabel("X axis")
ax.set_ylabel("Y axis")
ax.set_zlabel("Z axis")
plt.show()
```

```python
# Pandas for managing datasets
import pandas as pd

# Matplotlib for additional customization
from matplotlib import pyplot as plt
%matplotlib inline

# Seaborn for plotting and styling
import seaborn as sns

!wget -q https://elitedatascience.com/wp-content/uploads/2022/07/Pokemon.csv
# Read dataset
df = pd.read_csv('Pokemon.csv', index_col=0, encoding='latin')


# Display first 5 observations
df.head()
```

```python
# Scatterplot arguments
sns.lmplot(x='Attack', y='Defense', data=df,
           fit_reg=True, # No regression line
           hue='Stage')   # Color by evolution stage
```

```python
# Boxplot
plt.figure(figsize=(9,6)) # Set plot dimensions
sns.boxplot(data=df)
```

```python
# Preprocess DataFrame
stats_df = df.drop(['Total', 'Stage', 'Legendary'], axis=1)

# New boxplot using stats_df
plt.figure(figsize=(9,6)) # Set plot dimensions
sns.boxplot(data=stats_df)
```

```python
# Violin plot
plt.figure(figsize=(12,8)) # Set plot dimensions
sns.violinplot(x='Type 1', y='Attack', data=df)
```

```python
import seaborn as sns
import pandas as pd


features = ["HP", "Attack", "Defense", "Sp. Atk", "Sp. Def", "Speed"]
subset = df[features]

# Pearson correlation
pearson_corr = subset.corr(method='pearson')
# Spearman correlation
spearman_corr = subset.corr(method='spearman')

# Plot heatmaps with values displayed
plt.figure(figsize=(8,6))
sns.heatmap(pearson_corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Pearson Correlation Heatmap")
plt.show()

plt.figure(figsize=(8,6))
sns.heatmap(spearman_corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Spearman Correlation Heatmap")
plt.show()
```
